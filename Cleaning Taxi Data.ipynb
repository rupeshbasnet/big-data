```python
df["tpep_pickup_datetime"] = pd.to_datetime(df.tpep_pickup_datetime, format="%Y/%m/%d %H:%M:%S")
df["tpep_dropoff_datetime"] = pd.to_datetime(df.tpep_dropoff_datetime, format="%Y/%m/%d %H:%M:%S")
```

```python
# Adding the weekday
df['Index_Weekday'] = pd.to_datetime(df['tpep_pickup_datetime'])
days = {0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
df['Index_Weekday'] = df['Index_Weekday'].dt.dayofweek
df['Weekday'] = df['Index_Weekday'].apply(lambda x: days[x])
```

```python
df.head()
```

```python
df = df.drop(['VendorID'], axis=1)
df = df.drop(['trip_distance'], axis=1)
df = df.drop(['RatecodeID'], axis=1)
df = df.drop(['store_and_fwd_flag'], axis=1)
```

```python
df = df.drop(['payment_type', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge', 'total_amount'], axis=1)
```

```python
df = df.drop(['passenger_count'], axis=1)
```

```python
# Removing pickup long and lat with 0 removed
df = df[df.pickup_longitude != 0]
```

```python
dfAggValues = pd.DataFrame(df.Date.value_counts().sort_index().reset_index())
```

```python
dfAggValues.columns
```

```python
dfAggValues['Index_Weekday'] = pd.to_datetime(dfAggValues['Date'])
days = {0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
dfAggValues['Index_Weekday'] = dfAggValues['Index_Weekday'].dt.dayofweek
dfAggValues['Weekday'] = dfAggValues['Index_Weekday'].apply(lambda x: days[x])
```

```python
dfAggValues
```

```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (15, 8)
# Count of the pickups based on the month
dfAggValues.plot.bar(x='Date', y='Count')
```

```python
dfAggValues = dfAggValues.merge(dfWeather, on='Date', how='left')
dfAggValues.head()
```

```python
dfAvgValues = pd.DataFrame(dfAggValues.groupby('Weekday', as_index=False)['Count'].mean())
```

```python
dfAvgValues
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
mapDays = {day: i for i, day in enumerate(days)}
key = dfAvgValues['Weekday'].map(mapDays)
key
```

```python
dfAvgValues = dfAvgValues.iloc[key.argsort()].set_index('Weekday')
```
